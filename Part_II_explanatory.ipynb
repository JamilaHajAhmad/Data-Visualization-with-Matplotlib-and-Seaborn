## Investigation Overview

> **Summary of Main Findings:**
> - Customers tend to have longer trip durations compared to subscribers, likely due to different usage patterns (leisure vs. regular commutes).
> - Subscribers are more likely to participate in the bike share for all trips program compared to customers.
> - There is a noticeable difference in the distribution of `member_gender` between user types, with a higher proportion of male users among subscribers.
> - Male customers tend to have longer trip durations compared to female customers.
> - Subscribers who participate in the bike share for all trips program tend to have shorter trip durations, aligning with the idea of regular commutes.


## Dataset Overview and Executive Summary

> - The dataset consists of 183,412 rides with 16 features, including trip duration, start and end stations, start and end times, user type, and gender.

```python
# import all packages 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('./201902-fordgobike-tripdata.csv')
```

## Distribution of trip durations for customers and subscribers

> This visualization shows the distribution of trip durations for customers and subscribers. It highlights the differences in usage patterns between the two user types, with customers generally having longer trip durations compared to subscribers.

```python
plt.figure(figsize=(10, 6))

# Plotting the distribution of trip durations by user type
sns.histplot(data=df, x='duration_sec', hue='user_type', element='step', stat='density', common_norm=False)

# Adding title and labels
plt.title('Distribution of Trip Durations by User Type')
plt.xlabel('Trip Duration (seconds)')
plt.ylabel('Density')

# Limiting x-axis for better visualization
plt.xlim(0, 5000)

# Adding legend
plt.legend(title='User Type')

# Adding x and y ticks
plt.xticks(np.arange(0, 5001, 500))
plt.yticks(np.arange(0, 0.0011, 0.0001))

# Display the plot
plt.show()
```

## Distribution of Member Gender by User Type

> This visualization compares the distribution of `member_gender` between user types. It shows that a higher proportion of male users are subscribers compared to female users.

```python
plt.figure(figsize=(10, 6))

# Plotting the distribution of member gender by user type
sns.countplot(data=df, x='member_gender', hue='user_type', palette='viridis')

# Adding title and labels
plt.title('Distribution of Member Gender by User Type')
plt.xlabel('Member Gender')
plt.ylabel('Count')

# Adding legend
plt.legend(title='User Type')

# Display the plot
plt.show()
```

## Distribution of the average trip duration by user type

> This visualization shows the average trip duration for each `member_gender` grouped by `user_type`. It helps to identify if there is a significant difference in the trip durations between male and female users for subscribers and customers.

```python
# Calculate the average trip duration for each gender by user type
avg_trip_duration = df.groupby(['member_gender', 'user_type'])['duration_sec'].mean().reset_index()

# Plotting the average trip duration by user type
plt.figure(figsize=(10, 6))

sns.barplot(data=avg_trip_duration, x='member_gender', y='duration_sec', hue='user_type', palette='viridis')

# Adding title and labels
plt.title('Average Trip Duration by Member Gender and User Type')
plt.xlabel('Member Gender')
plt.ylabel('Average Trip Duration (seconds)')

# Adding legend
plt.legend(title='User Type')

# Display the plot
plt.show()
```
